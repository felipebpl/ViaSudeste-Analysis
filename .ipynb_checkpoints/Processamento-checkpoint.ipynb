### Importando as bibliotecas 

```python
import pandas as pd
import numpy as np
```

### Base de dados

```python
df_pedidos = pd.read_excel("data/pedidos.xlsx")
df_itens = pd.read_excel("data/itens.xlsx") 
df_clientes = pd.read_excel("data/clientes.xlsx") 
```

Agrupando os dados de clientes

```python
pedidos_clientes = df_clientes.merge(df_pedidos, how="inner", on="cliente_id")
pedidos_clientes_itens = pedidos_clientes.merge(df_itens, how="inner", on="pedido_id")

# Filtrando as colunas desejadas
pedidos_clientes_itens.drop(columns=[i for i in pedidos_clientes_itens if i not in ['cliente_id_unico', 'pedido_id', 'pedido_status','preco', 'preco_frete']],inplace=True  )
# pedidos_clientes_itens = pedidos_clientes_itens[['cliente_id_unico', 'pedido_id', 'pedido_status','preco', 'preco_frete']]
```

```python
pedidos_clientes_itens.pedido_status.unique()
```

```python
# filtrando os registros de pedidos indisponíveis ou cancelados
pedidos_clientes_itens = pedidos_clientes_itens[(pedidos_clientes_itens.pedido_status != 'cancelado') & (pedidos_clientes_itens.pedido_status != 'indisponivel')]
```

```python
valor_por_cliente = pedidos_clientes_itens.groupby(by='cliente_id_unico').sum()
```

```python
valor_por_cliente["valor_total"] = valor_por_cliente["preco"] + valor_por_cliente["preco_frete"]
```

```python
valor_por_cliente.describe()
```

```python
# filtrando os registros de pedidos indisponíveis ou cancelados
pedidos_clientes = pedidos_clientes[(pedidos_clientes.pedido_status != 'cancelado') & (pedidos_clientes.pedido_status != 'indisponivel') & (pedidos_clientes.pedido_status != 'created')]

qnt_por_clientes = pedidos_clientes.groupby(by="cliente_id_unico").count()["pedido_id"]
 
qnt_por_clientes = qnt_por_clientes.reset_index(level=0)
valor_por_cliente = valor_por_cliente.reset_index(level=0)
```

```python
# valor_por_cliente = valor_por_cliente.reset_index(level=0)
```

```python
info_clientes = valor_por_cliente.merge(qnt_por_clientes, how="inner", on="cliente_id_unico")
info_clientes.drop(columns=['preco', 'preco_frete'], inplace=True)
info_clientes.rename(columns={'pedido_id': 'qnt_pedidos'}, inplace=True)
info_clientes.describe()
```
